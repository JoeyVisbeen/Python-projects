````python
from elasticsearch import Elasticsearch
import pandas as pd

es = Elasticsearch('http://localhost:9200')
````

````python
df_categories = (
    pd.read_csv('../amazon_categories.csv')
    .dropna()
    .reset_index()
)
df_categories
````

````python
df_products = (
    pd.read_csv('../amazon_products.csv')
    .dropna()
    .sample(5000, random_state=42)
    .reset_index()
)
````

````python
df_products
````

Mergen en vervangen van de category id's met de namen van de categorieen.  

````python
merged_df = pd.merge(df_products, df_categories, left_on='category_id', right_on='id', how='left')
merged_df.drop('category_id', axis=1)
````

````python
mappings = {
    'properties': {
        'title': {'type': 'text'},
        'imageUrl': {'type': 'text'},
        'productUrl': {'type': 'text'},
        'stars': {'type': 'integer'},
        'reviews': {'type': 'integer'},
        'price': {'type': 'double'},
        'listPrice': {'type': 'double'},
        'categoryName': {'type': 'text'},
        'isBestSeller': {'type': 'boolean'},
        'boughtLastMonth': {'type': 'integer'}
    }
}

es.indices.create(index='amazon_products', mappings=mappings)
#es.indices.delete(index='amazon_products')
````

````python
from elasticsearch.helpers import bulk

bulk_data = []

for i,row in merged_df.iterrows():
    bulk_data.append({
        '_index': 'amazon_products',
        '_id': i,
        '_source': {
            'title': row['title'],
            'imageUrl': row['imgUrl'],
            'productUrl': row['productURL'],
            'stars': row['stars'],
            'reviews': row['reviews'],
            'price': row['price'],
            'listPrice': row['listPrice'],
            'categoryName': row['category_name'],
            'isBestSeller': row['isBestSeller'],
            'boughtLastMonth': row['boughtInLastMonth']
        }
    })
#bulk_data
bulk(es, bulk_data)    
````

````python
es.indices.refresh(index='amazon_products')
````

````python
es.cat.count(index='amazon_products', format='json')
````

Query's uitvoeren

````python
resp = es.search( index='amazon_products')
resp
````

````python
query = {
    'match': {
        'title': 'xbox games'        
    }
}

result = es.search(
    index='amazon_products', 
    query=query)

result['hits']['hits']
````
